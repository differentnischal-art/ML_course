```python
def convert_size(val):
    if isinstance(val,str):#checks is it string or not
        val=val.strip()#removes whitespaces if there is any
        if val.endswith('M'):
            return float(val[:-1])*1000000#converts all the elements except the last i.e by -1 and multiplies 
        elif val.endswith('k'):
            return float(val[:-1])*1000
        else:
            return None
df_copy['Size']=df['Size'].apply(convert_size)#by apply function passing every value to the user defined function
```

```python
def remove_string(xyz):
    if isinstance(xyz,str):
        if xyz.endswith('+'):
            return float (xyz[:-1].replace(',',''))
    return None
df_copy['Installs']=df_copy['Installs'].apply(remove_string)
```
