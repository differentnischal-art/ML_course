```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,weights=[0.76],random_state=12)
```

```python
import pandas as pd 
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])#c means colors based on target value i.e either for 0 or 1
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
X,y=oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```

```python
X.shape
```

```python
y.shape
```

```python
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
oversample_df=pd.concat([df1,df2],axis=1)
```

```python
plt.scatter(oversample_df['f1'],oversample_df['f2'],c=oversample_df['target'])
```
