```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
df=pd.DataFrame({'color':['red','blue','green','red','blue']})
df.head()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

ORDINAL ENCODING


```python
df=pd.DataFrame({'rank':['good','better','best','best','good','good','better']})
```

```python
df.tail()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder=OrdinalEncoder(categories=[['good','better','best']])
```

```python
encoder.fit_transform(df[['rank']])
```

Target GUIDED ORDINAL Encoding

```python
import seaborn as sns
df1=sns.load_dataset('tips')
```

```python
df1.head()
```

```python
df1['time']
```

```python
df1.info()
```

```python
time_order=df1.groupby('time')['tip'].mean().sort_values()
```

```python
ordinal_map = {k:i for i,k in enumerate(time_order)}
ordinal_map
```

```python
df1['time_encoded'] = df1['time'].map(ordinal_map)
```
