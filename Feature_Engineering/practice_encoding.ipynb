```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder=OneHotEncoder()
```

```python
df=pd.DataFrame({
    'color':['red','blue','green','orange','yellow']
})
```

```python
df.head()
```

```python
encoded=encoder.fit_transform(df[['color']]).toarray() #firstly these tuples are sorted in alphabetical  order then it is given the index and represented in the form matrix
```

```python
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoder_df.head()
```
